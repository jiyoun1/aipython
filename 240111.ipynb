```python
# 붓꽃 데이터 분석

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('font', family='NanumGothic')
df = pd.read_csv('iris.csv')

plt.bar(df['species'],df['petal_length'])
plt.title('품종별 sepal_length 그래프')
plt.xlabel('종류')
plt.ylabel('꽃받침 길이의 평균값')
plt.show()

plt.hist(df['sepal_length'],bins= 3, label = 'sepal_length')     # 각 속성값들의 빈도수 히스토그램
plt.hist(df['sepal_width'], bins = 3, label = 'sepal_width')
plt.hist(df['petal_length'], bins = 3, label = 'petal_length')
plt.hist(df['petal_width'], bins = 3, label = 'petal_width')
plt.title('각 속성값들의 빈도수 히스토그램')
plt.xlabel('value')
plt.ylabel('빈도수')
plt.legend()
plt.show()

plt.scatter(df['sepal_length'], df['petal_length'])            #꽃받침의 길이와 꽃잎의 길이의 산점도
plt.xlabel('sepal_length')
plt.ylabel('petal_length')

plt.show()
```

```python
# 타이타닉 데이터 분석

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('font', family='NanumGothic')
df = pd.read_csv('titanic.csv')

df.info() #기본 정보 출력
df.describe()  #기초 통계량 출력
df.sort_values(by=['fare'],ascending = False)  #요금 기준 내림차순 정렬
df.groupby('survived').count()

df.isna().sum()    #결측치 확인
df1 = df.drop('deck',axis=1)  #결측치 컬럼(deck) 삭제

df1['embarked'].value_counts() #탑승항구 최빈값 확인
df2 = df1.fillna({'embarked':'S'}) #탑승항구 결측치 최빈값 대체

age_mean = df2['age'].mean()     #나이 평균
df3 = df2.fillna({'age':age_mean})  #나이 결측치 평균 대체

emb = df3.groupby('embarked')   #항구별 pclass컬럼의 평균값 출력
emb['pclass'].mean()

pcl = df3.groupby('pclass')

#sex = pcl.groupby('sex')
#sex.mean()

# titanic=df3.corr()
# sns.heatmap(data=titanic,annot=True,fmt='.3f',cbar=False)
# plt.show()
```
