```python
# 붓꽃 데이터 분석

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('font', family='NanumGothic')
df = pd.read_csv('iris.csv')

df.info() #기본 정보 출력

df.describe()  #기초 통계량 출력

df['species'].value_counts()  #품종별 개수

df.isna().sum()

df.duplicated()

df.drop_duplicates()

df.groupby('species').sum()

df.groupby('species').mean()
```

```python
# 붓꽃 데이터 분석

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rc('font', family='NanumGothic')
df = pd.read_csv('iris.csv')

sns.barplot(x= 'species', y='sepal_length',data = df,color = 'b',errorbar = None)
plt.title('품종별 sepal_length 그래프')
plt.xlabel('종류')
plt.ylabel('꽃받침 길이의 평균값')
plt.show()

plt.hist(df['sepal_length'],bins= 3, label = 'sepal_length')     # 각 속성값들의 빈도수 히스토그램
plt.hist(df['sepal_width'], bins = 3, label = 'sepal_width')
plt.hist(df['petal_length'], bins = 3, label = 'petal_length')
plt.hist(df['petal_width'], bins = 3, label = 'petal_width')
plt.title('각 속성값들의 빈도수 히스토그램')
plt.xlabel('value')
plt.ylabel('빈도수')
plt.legend()
plt.show()

plt.scatter(df['sepal_length'], df['petal_length'])            #꽃받침의 길이와 꽃잎의 길이의 산점도
plt.xlabel('sepal_length')
plt.ylabel('petal_length')

plt.show()
```
